```python
import sys
import os
from vosk import Model, KaldiRecognizer
import pyaudio
import time
import json
from mistralai import Mistral
import pyttsx3
import threading
import tkinter as tk
from tkinter import font

# ====================== GUI Setup ======================

# Create the main window
root = tk.Tk()
root.title("Борис Ельцин")
root.geometry("400x300")  # Set window size
root.configure(bg="white")  # Initial background color (idle state)

# Create a custom font for the status label
status_font = font.Font(family="Helvetica", size=20, weight="bold")

# Create a label to display the status
status_label = tk.Label(
    root,
    text="Слушаю...",
    font=status_font,
    bg="white",
    fg="black",
)
status_label.pack(expand=True, fill="both")  # Make the label fill the window

# ====================== Speech Recognition and TTS ======================

# Load the Vosk model
model = Model("vosk-model-small-ru-0.22")

# Initialize the recognizer
recognizer = KaldiRecognizer(model, 16000)

# Initialize the TTS engine
engine = pyttsx3.init()
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[0].id)
engine.setProperty('voice', 'ru')

# Function to update the GUI state
def update_gui_state(state):
    if state == "idle":
        root.configure(bg="white")
        status_label.config(text="Слушаю...", bg="white", fg="black")
    elif state == "triggered":
        root.configure(bg="yellow")
        status_label.config(text="Слушаю ваш вопрос...", bg="yellow", fg="black")
    elif state == "speaking":
        root.configure(bg="red")
        status_label.config(text="Говорю...", bg="red", fg="white")
    root.update()  # Update the GUI

# Function to listen continuously
def listen_continuously():
    # Start audio stream
    mic = pyaudio.PyAudio()
    stream = mic.open(format=pyaudio.paInt16, channels=1, rate=16000, input=True, frames_per_buffer=8000, input_device_index=5)
    stream.start_stream()

    print("Listening for the trigger word 'борис'...")

    while True:
        # Read audio data from the stream
        data = stream.read(4000, exception_on_overflow=False)

        # Process the audio data
        if recognizer.AcceptWaveform(data):
            # Get the final result
            result = recognizer.Result()
            result_json = json.loads(result)
            recognized_text = result_json.get("text", "").lower()  # Convert to lowercase for easier matching

            # Check if the trigger word "борис" is detected
            if "борис" in recognized_text:
                print("Trigger word 'борис' detected. Listening for user message...")
                update_gui_state("triggered")  # Change to yellow (triggered state)
                user_message = get_user_message(stream)  # Capture the user message
                if user_message:
                    response = run_mistral(user_message)  # Generate response using Mistral AI
                    update_gui_state("speaking")  # Change to red (speaking state)
                    speak_response(response)  # Speak the response
                else:
                    print("No speech detected after trigger.")
                update_gui_state("idle")  # Return to white (idle state)
        else:
            # Get the partial result (optional, for debugging)
            partial_result = recognizer.PartialResult()
            print("Partial Result:", partial_result)

# Function to capture the user message after the trigger word
def get_user_message(stream):
    user_message = ""
    start_time = time.time()

    while True:
        # Check if 8 seconds have passed
        if time.time() - start_time >= 8:
            print("Time limit reached. Stopping user message capture.")
            break

        # Read audio data from the stream
        data = stream.read(4000, exception_on_overflow=False)

        # Process the audio data
        if recognizer.AcceptWaveform(data):
            # Get the final result
            result = recognizer.Result()
            result_json = json.loads(result)
            user_message = result_json.get("text", "")
            print("User Message:", user_message)
            break
        else:
            # Get the partial result (optional, for debugging)
            partial_result = recognizer.PartialResult()
            print("Partial Result:", partial_result)

    return user_message

# Function to run Mistral AI
def run_mistral(user_message, model="mistral-large-latest"):
    # Initialize the Mistral client
    api_key = "Rsha4RoCjdjybQ6bmtMrTJPRkavjgPfe"
    client = Mistral(api_key=api_key)
    
    # Define the system message and user message
    messages = [
        {"role": "system", "content": "Ты первый президент Российской Федерации Борис Ельцин. Веди себя как агрессивная пьяница и используй цитаты Ельцина при ответах на вопросы. Отвечай не более чем двумя предложениями. Отвечай агрессивно, иногда угрожай. Периодически матерись."},
        {"role": "user", "content": user_message}
    ]
    
    # Get the chat response
    chat_response = client.chat.complete(
        model=model,
        messages=messages
    )
    
    # Return the assistant's response
    return chat_response.choices[0].message.content

# Function to speak the response
def speak_response(response):
    print("Speaking response:", response)
    engine.say(response)
    engine.runAndWait()

# Start the continuous listening in a separate thread
threading.Thread(target=listen_continuously, daemon=True).start()

# Start the GUI event loop
root.mainloop()
```
