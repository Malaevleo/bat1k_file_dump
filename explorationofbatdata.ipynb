```python
from Bio import SeqIO
import pandas as pd

# File paths
matrisome_fasta_file = "homo_collagens.fasta"
bat_fasta_file = "modified_bat_proteins.faa"
blast_results_file = "realmergedcollagenssearch.txt"  # Should contain the best_matches DataFrame
output_file = "realmergedcollaganessearchresults.fasta"

# Function to load sequences while handling duplicates
def load_fasta_with_duplicates(fasta_file):
    sequences = {}
    for record in SeqIO.parse(fasta_file, "fasta"):
        if record.id not in sequences:
            sequences[record.id] = record
        else:
            print(f"Duplicate ID found: {record.id}, ignoring this entry.")
    return sequences

# Load sequences from both FASTA files
matrisome_sequences = load_fasta_with_duplicates(matrisome_fasta_file)
bat_sequences = load_fasta_with_duplicates(bat_fasta_file)

# Load best_matches DataFrame
blast_df = pd.read_csv(blast_results_file, sep="\t")

# Open the output FASTA file
with open(output_file, "w") as output_fasta:
    for qseqid, group in blast_df.groupby("qseqid"):
        # Retrieve and write the human protein sequence
        if qseqid in matrisome_sequences:
            qseq = matrisome_sequences[qseqid].seq
            qdesc = matrisome_sequences[qseqid].description
            output_fasta.write(f">{qseqid} {qdesc}\n{qseq}\n")
        else:
            print(f"QseqID {qseqid} not found in human matrisome.")
            continue

        # Write each aligned bat protein sequence for the current human protein
        for _, row in group.iterrows():
            sseqid_full = row["sseqid"]
            prefix = sseqid_full[:2]
            sseqid = sseqid_full[2:]  # Remove the prefix
            
            if sseqid in bat_sequences:
                sseq = bat_sequences[sseqid].seq
                sdesc = bat_sequences[sseqid].description
                
                # Note the ID source based on prefix
                if prefix == "pn" or prefix == 'en':
                    id_source = "GenBank"
                else:
                    id_source = "RefSeq"
                
                # Write the bat protein sequence
                output_fasta.write(f">{sseqid} {sdesc} [{id_source}]\n{sseq}\n")
            else:
                print(f"SseqID {sseqid} not found in bat proteome.")

print(f"Paired sequences written to {output_file}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Load your data into a DataFrame
data = pd.read_csv("aggregated_clusters.tsv", sep="\t")

# Define colors based on prefixes
color_mapping = {
    'green': {'sb', 'pn', 'pk', 'mo', 'pd', 'pv', 'pa'},
    'yellow': {'en', 'ef', 'ph', 'aj', 'ra', 'pg'},
    'red': {'mb', 'md', 'ml', 'mm', 'dr', 'rf'}
}

# Function to assign color based on prefix
def get_color(prefix):
    for color, prefixes in color_mapping.items():
        if prefix in prefixes:
            return color
    return 'gray'  # Default color if prefix is not found

# Initialize the main graph
G = nx.Graph()

# Build the graph by adding clusters
for _, row in data.iterrows():
    cluster_members = eval(row["member"])  # Evaluate the string as a list
    representative = row["representative"]
    
    # Add nodes and edges for the cluster
    for member in cluster_members:
        prefix = member[:2]  # Extract the prefix
        color = get_color(prefix)  # Get color based on prefix
        G.add_node(member, color=color)
        
    # Connect all members within the cluster
    for i in range(len(cluster_members)):
        for j in range(i + 1, len(cluster_members)):
            G.add_edge(cluster_members[i], cluster_members[j])

# Get colors for all nodes
node_colors = [G.nodes[node]['color'] for node in G.nodes]

# Plot the graph
plt.figure(figsize=(20, 20))
pos = nx.spring_layout(G, seed=42)  # Layout for better visualization
nx.draw(G, pos, node_color=node_colors, with_labels=True, node_size=50, font_size=8, edge_color="gray", alpha=0.6)
plt.title("Protein Clusters Visualization")
plt.show()
```

```python
short_living_prefixes = {'sb', 'pn', 'pk', 'mo', 'pd', 'pv', 'pa'}
average_living_prefixes = {'en', 'ef', 'ph', 'aj', 'ra', 'pg'}
long_living_prefixes = {'mb', 'md', 'ml', 'mm', 'dr', 'rf'}

# Function to categorize a prefix
def get_longevity_category(member_id):
    prefix = member_id[:2]  # Get the first two letters as prefix
    if prefix in short_living_prefixes:
        return 'short'
    elif prefix in average_living_prefixes:
        return 'average'
    elif prefix in long_living_prefixes:
        return 'long'
    else:
        return 'unknown'

# Process each cluster and calculate longevity association
longevity_clusters = []

for _, row in aggregated_df3.iterrows():
    members = row['member']  # List of members in the cluster
    
    # Skip clusters with only 1 or 2 proteins
    if len(members) <= 4:
        continue
    
    # Initialize count for each longevity category
    longevity_counts = {'short': 0, 'average': 0, 'long': 0}
    
    # Count each member's longevity category
    for member_id in members:
        category = get_longevity_category(member_id)
        if category in longevity_counts:
            longevity_counts[category] += 1
    
    # Calculate the percentage of short-living or long-living bats
    total_members = len(members)
    short_percentage = longevity_counts['short'] / total_members
    long_percentage = longevity_counts['long'] / total_members
    
    # Check if more than 75% of the cluster members are short or long-living
    if short_percentage > 0.75:
        longevity_clusters.append((row['representative'], 'short', short_percentage))
    elif long_percentage > 0.75:
        longevity_clusters.append((row['representative'], 'long', long_percentage))

longevity_clusters_df = pd.DataFrame(longevity_clusters, columns=['representative', 'dominant_longevity', 'percentage'])

print(longevity_clusters_df)
```

```python
import pandas as pd

myotis_prefixes = {'mm', 'ml', 'md', 'mb'}
pteropus_prefixes = {'pv', 'pg', 'pa'}
phyllostomus_prefixes = {'ph', 'pd'}
pipistrellus_prefixes = {'pk', 'pn'}
eptesicus_prefixes = {'en', 'ef'}

def get_genus(prefix):
    if prefix in myotis_prefixes:
        return 'Myotis'
    elif prefix in pteropus_prefixes:
        return 'Pteropus'
    elif prefix in phyllostomus_prefixes:
        return 'Phyllostomus'
    elif prefix in pipistrellus_prefixes:
        return 'Pipistrellus'
    elif prefix in eptesicus_prefixes:
        return 'Eptesicus'
    else:
        return 'Other'


phylogenetic_clusters = []

for _, row in aggregated_df3.iterrows():
    members = row['member']  # List of members in the cluster
    genus_count = {}
    species_count = set()  # Track distinct species by prefix in each genus

    # Count members per genus and species per genus
    for member_id in members:
        prefix = member_id[:2]
        genus = get_genus(prefix)
        if genus != 'Other':
            genus_count[genus] = genus_count.get(genus, 0) + 1
            species_count.add(prefix)  # Track species by prefix for diversity

    # Check if cluster has proteins only from one genus with at least 2 different species
    for genus, count in genus_count.items():
        if len(species_count) > 1 and count == len(members):  # All proteins belong to the same genus with diversity
            phylogenetic_clusters.append((row['representative'], genus, list(species_count), members))


phylogenetic_clusters_df = pd.DataFrame(phylogenetic_clusters, columns=['representative', 'genus', 'species_in_genus', 'full_members'])

print(phylogenetic_clusters_df)
```

```python
import pandas as pd
rbhbigcl_df = pd.read_csv("rbhclustersv5.tsv", sep="\t", header=None, names=["representative", "member"])

aggregated_df4 = rbhbigcl_df.groupby("representative")["member"].apply(list).reset_index()

print(aggregated_df4.head())

aggregated_df4.to_csv("aggregated_clusters_bigclust.tsv", sep="\t", index=False)
```

```python
short_living_prefixes = {'sb', 'pn', 'pk', 'mo', 'pd', 'pv', 'pa'}
average_living_prefixes = {'en', 'ef', 'ph', 'aj', 'ra', 'pg'}
long_living_prefixes = {'mb', 'md', 'ml', 'mm', 'dr', 'rf'}

# Function to categorize a prefix
def get_longevity_category(member_id):
    prefix = member_id[:2]  # Get the first two letters as prefix
    if prefix in short_living_prefixes:
        return 'short'
    elif prefix in average_living_prefixes:
        return 'average'
    elif prefix in long_living_prefixes:
        return 'long'
    else:
        return 'unknown'

# Process each cluster and calculate longevity association
longevity_clusters = []

for _, row in aggregated_df4.iterrows():
    members = row['member']  # List of members in the cluster
    
    # Skip clusters with only 1 or 2 proteins
    if len(members) <= 4:
        continue
    
    # Initialize count for each longevity category
    longevity_counts = {'short': 0, 'average': 0, 'long': 0}
    
    # Count each member's longevity category
    for member_id in members:
        category = get_longevity_category(member_id)
        if category in longevity_counts:
            longevity_counts[category] += 1
    
    # Calculate the percentage of short-living or long-living bats
    total_members = len(members)
    short_percentage = longevity_counts['short'] / total_members
    long_percentage = longevity_counts['long'] / total_members
    
    # Check if more than 75% of the cluster members are short or long-living
    if short_percentage > 0.75:
        longevity_clusters.append((row['representative'], 'short', short_percentage))
    elif long_percentage > 0.75:
        longevity_clusters.append((row['representative'], 'long', long_percentage))

# Convert to DataFrame for easier viewing
longevity_clusters_df = pd.DataFrame(longevity_clusters, columns=['representative', 'dominant_longevity', 'percentage'])

# Display the result
print(longevity_clusters_df)
```
