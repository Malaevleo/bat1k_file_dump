```python
from Bio import Entrez
from http.client import IncompleteRead
import time


# Input file containing gene names
gene_file = "genes_by_cat_lost/transcriptional_alterations_genes.txt"

# Output FASTA file for large genes (protein sequences)
output_fasta = "genes_by_cat_lost/transcriptional_alterations_genes_proteins.fasta"

# Function to fetch sequences robustly with retries
def robust_fetch(fetch_function, retries=3, wait_time=5):
    for attempt in range(retries):
        try:
            return fetch_function()
        except IncompleteRead as e:
            print(f"IncompleteRead error: {e}. Retrying {attempt + 1}/{retries}...")
            time.sleep(wait_time)
    raise Exception("Failed to fetch data after multiple retries")

# Fetch protein sequences from NCBI
def fetch_protein_sequence(gene_name):
    try:
        # Search for the gene in the NCBI protein database (restricted to Homo sapiens)
        query = f"{gene_name}[Gene] AND Homo sapiens[Organism]"
        handle = Entrez.esearch(db="protein", term=query, retmax=1)
        record = Entrez.read(handle)
        handle.close()

        # Fetch the protein sequence by ID
        if record["IdList"]:
            seq_id = record["IdList"][0]
            print(f"Fetching protein sequence for {gene_name} (ID: {seq_id})...")
            handle = robust_fetch(
                lambda: Entrez.efetch(
                    db="protein", id=seq_id, rettype="fasta", retmode="text"
                )
            )
            sequence = handle.read()
            handle.close()
            return sequence
        else:
            print(f"No protein sequence found for {gene_name}")
            return None
    except Exception as e:
        print(f"Error fetching {gene_name}: {e}")
        return None

# Read gene names and fetch protein sequences
with open(gene_file, "r") as file:
    genes = [line.strip() for line in file if line.strip()]

with open(output_fasta, "w") as fasta_out:
    for gene in genes:
        protein_sequence = fetch_protein_sequence(gene)
        if protein_sequence:
            fasta_out.write(protein_sequence)

print(f"Protein sequences saved to: {output_fasta}")
```

```python
import requests
import os
import glob

# Directory containing gene list TXT files
directory = "genes_by_cat_lost"  

def fetch_uniprot_sequences(gene_names):
    base_url = "https://rest.uniprot.org/uniprotkb/search"
    headers = {"Accept": "text/x-fasta"}
    sequences = {}

    for gene_name in gene_names:
        print(f"Fetching sequence for: {gene_name}")
        query = f"gene_exact:{gene_name} AND organism:9606"
        params = {"query": query, "format": "fasta"}
        response = requests.get(base_url, headers=headers, params=params)
        
        if response.status_code == 200 and response.text.strip():
            sequences[gene_name] = response.text.strip()
        else:
            print(f"No sequence found for {gene_name} or error {response.status_code}")
    
    return sequences

# Find all TXT files in the directory
txt_files = glob.glob(os.path.join(directory, "*.txt"))

# Process each TXT file separately
for gene_file in txt_files:
    base_name = os.path.splitext(os.path.basename(gene_file))[0]
    output_fasta = os.path.join(directory, base_name + "_proteins.fasta")
    print(f"\nProcessing file: {gene_file}")

    # Read gene names from the current file
    with open(gene_file, "r") as file:
        genes = [line.strip() for line in file if line.strip()]

    # Fetch sequences and save to a FASTA file
    sequences = fetch_uniprot_sequences(genes)
    
    with open(output_fasta, "w") as fasta_out:
        for seq in sequences.values():
            fasta_out.write(seq + "\n")

    print(f"Protein sequences saved to: {output_fasta}")
```

```python
from biomart import BiomartServer
import os
import glob

# Directory containing gene TXT files
directory = "genes_by_cat_lost"

server = BiomartServer("http://www.ensembl.org/biomart")
dataset = server.datasets['hsapiens_gene_ensembl']  # Homo sapiens dataset

def fetch_protein_sequences(gene_names):
    sequences = {}
    for gene_name in gene_names:
        print(f"Fetching sequence for: {gene_name}")
        result = dataset.search({
            'filters': {
                'hgnc_symbol': gene_name
            },
            'attributes': [
                'hgnc_symbol',  # Human Gene Symbol
                'protein_sequence'  # Correct attribute for amino acid sequences
            ]
        })

        output = result.text.strip().split("\n")
        if output and len(output[0].strip()) > 0:
            sequences[gene_name] = output
        else:
            print(f"No sequence found for {gene_name}")
    return sequences

# Process each TXT file separately
txt_files = glob.glob(os.path.join(directory, "*.txt"))

for gene_file in txt_files:
    base_name = os.path.splitext(os.path.basename(gene_file))[0]
    output_fasta = os.path.join(directory, base_name + "_biomart_proteins.fasta")
    print(f"\nProcessing file: {gene_file}")

    # Read gene names from the current file
    with open(gene_file, "r") as file:
        genes = [line.strip() for line in file if line.strip()]

    # Fetch sequences and save to a FASTA file
    sequences = fetch_protein_sequences(genes)

    with open(output_fasta, "w") as fasta_out:
        for gene, seq_list in sequences.items():
            for seq in seq_list:
                fasta_out.write(f">{gene}\n{seq.strip()}\n")

    print(f"Protein sequences saved to: {output_fasta}")
```

```python
import requests
import os
import glob

# Directory containing gene TXT files
directory = "genes_by_cat_lost/redo"

# Base URL for UniProt batch retrieval
uniprot_url = "https://rest.uniprot.org/idmapping/run"
retrieve_url = "https://rest.uniprot.org/idmapping/results/"

def get_uniprot_ids(gene_list):
    """Submit genes to UniProt batch API to get mapping results."""
    headers = {"Content-Type": "application/json"}
    payload = {
        "from": "Gene_Name",
        "to": "UniProtKB",
        "format": "fasta",
        "ids": gene_list
    }

    # Submit the request
    response = requests.post(uniprot_url, json=payload, headers=headers)
    if response.status_code == 200:
        job_id = response.json().get("jobId")
        print(f"Job ID: {job_id}")
        return job_id
    else:
        print(f"Failed to submit job: {response.status_code}")
        return None

def retrieve_fasta(job_id):
    """Retrieve FASTA results using the job ID."""
    for _ in range(20):  # Retry logic if data isn't immediately available
        result_response = requests.get(f"{retrieve_url}{job_id}")
        if result_response.status_code == 200:
            return result_response.text
        elif result_response.status_code == 404:
            continue
    print(f"Failed to retrieve job {job_id}")
    return ""

def process_txt_files():
    """Process all gene text files and fetch amino acid sequences."""
    txt_files = glob.glob(os.path.join(directory, "*.txt"))

    for gene_file in txt_files:
        # Read gene names from the file
        base_name = os.path.splitext(os.path.basename(gene_file))[0]
        output_fasta = os.path.join(directory, base_name + "_uniprot_proteins.fasta")

        print(f"\nProcessing file: {gene_file}")
        with open(gene_file, "r") as file:
            genes = [line.strip() for line in file if line.strip()]

        # Submit to UniProt and retrieve sequences
        job_id = get_uniprot_ids(genes)
        if job_id:
            fasta_data = retrieve_fasta(job_id)
            if fasta_data:
                with open(output_fasta, "w") as fasta_out:
                    fasta_out.write(fasta_data)
                print(f"Protein sequences saved to: {output_fasta}")
            else:
                print(f"No sequences retrieved for file: {gene_file}")

process_txt_files()
```
