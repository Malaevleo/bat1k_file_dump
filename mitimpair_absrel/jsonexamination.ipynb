```python
import os
import json
import pandas as pd

# Specify the folder containing the JSON files
folder_path = "mitimpair_absrel"

# Initialize a dictionary to store DataFrames
dataframes = {}

# Iterate over all JSON files in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.json'):
        file_path = os.path.join(folder_path, file_name)
        try:
            # Load the JSON data
            with open(file_path, 'r') as f:
                json_data = json.load(f)
            
            # Extract relevant data for each tested node
            branch_attributes = json_data.get("branch attributes", {}).get("0", {})
            if branch_attributes:
                df_data = {node: {
                    "Baseline MG94xREV omega ratio": attr.get("Baseline MG94xREV omega ratio", None),
                    "Rate classes": attr.get("Rate classes", None),
                    "Rate distributions": attr.get("Rate Distributions", []),
                    "Corrected P-value": attr.get("Corrected P-value", None),
                } for node, attr in branch_attributes.items()}
                
                # Create a DataFrame from the extracted data
                df = pd.DataFrame.from_dict(df_data, orient="index")
                dataframes[file_name] = df

        except (json.JSONDecodeError, FileNotFoundError) as e:
            print(f"Error processing file {file_name}: {e}")
            continue

# Display the DataFrame for each JSON file
for file_name, df in dataframes.items():
    print(f"DataFrame for {file_name}:")
    print(df)
    print("\n")

# Save DataFrames to CSV if needed
output_folder = "output_csvs"
os.makedirs(output_folder, exist_ok=True)
for file_name, df in dataframes.items():
    output_file = os.path.join(output_folder, file_name.replace('.json', '.csv'))
    df.to_csv(output_file, index=True)
```
